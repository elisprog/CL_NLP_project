```python
import requests
import re
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
UserAgent().chrome
```

```python
inicial = 'https://cvc.cervantes.es/ensenanza/biblioteca_ele/plan_curricular/niveles/09_nociones_especificas_inventario_a1-a2.htm'
medium = 'https://cvc.cervantes.es/ensenanza/biblioteca_ele/plan_curricular/niveles/09_nociones_especificas_inventario_b1-b2.htm'
advanced = 'https://cvc.cervantes.es/ensenanza/biblioteca_ele/plan_curricular/niveles/09_nociones_especificas_inventario_c1-c2.htm'
cervantes_links = [inicial, medium, advanced]
words_levels = []
for link in cervantes_links:
    link_page = requests.get(link, headers={'User-Agent': UserAgent().chrome})
    big_page_soup = BeautifulSoup(link_page.text)
    big_page_soup_links = big_page_soup.find_all('li')
    for each in big_page_soup_links:
        words_levels.append(each.get_text())
```

```python
with open('words_cervantes.txt', 'w', encoding='UTF-16LE') as file:
    file.write(' '.join(words_levels))
```

```python
with open('words_cervantes.txt', 'r', encoding='UTF-16LE') as file:
    words_cervantes = file.read()
    
words_cervantes = re.sub('v. ', '', words_cervantes.lower())
words_cervantes = re.sub('[~,.\[\]/\(\)0123456789+¡!¿?]', ' ', words_cervantes)
words_cervantes = re.sub('ã¡', 'á', words_cervantes)
words_cervantes = re.sub('ã±', 'ñ', words_cervantes)
words_cervantes = re.sub('ã©', 'é', words_cervantes)
words_cervantes = re.sub('ã', 'í', words_cervantes)
words_cervantes = re.sub('ã³', 'ó', words_cervantes)
words_cervantes = re.sub('í³', 'ó', words_cervantes)
words_cervantes = re.sub('íº', 'ú', words_cervantes)
words_cervantes = re.sub('ã¼', 'ü', words_cervantes)
words_cervantes = re.sub('ejemplo propuesto ', '', words_cervantes)
words_cervantes = re.sub(' la ', '', words_cervantes)
words_cervantes = re.sub(' las ', '', words_cervantes)
words_cervantes = re.sub(' el ', '', words_cervantes)
words_cervantes = re.sub(' los ', '', words_cervantes)
words_cervantes = re.sub(' un ', '', words_cervantes)
words_cervantes = re.sub(' una ', '', words_cervantes)
words_cervantes = re.sub(' de ', '', words_cervantes)
words_cervantes = re.sub(' del ', '', words_cervantes)
words_cervantes = re.sub(' en ', '', words_cervantes)
words_cervantes = re.sub(' a ', '', words_cervantes)
words_cervantes = re.sub(' al ', '', words_cervantes)
words_cervantes = re.sub(' se ', '', words_cervantes)
words_cervantes = re.sub(' in ', '', words_cervantes)
words_cervantes = re.sub(' im ', '', words_cervantes)
words_cervantes = re.sub(' no ', '', words_cervantes)
words_cervantes = re.sub('â', '', words_cervantes)
words_cervantes = re.sub('\xad', '', words_cervantes)

with open('words_cervantes.txt', 'w', encoding='UTF-16LE') as file:
    file.write(words_cervantes)
    
print(len(words_cervantes.split()),words_cervantes.split())
```
